```python
alpha = dict()
asc = ord('a')
for i in range(26):
  alpha.update({i+1 : chr(asc+i)})
```

```python
#define the number of samples for training set and for validation set
num_train,num_validation = int(len(train_file)*0.9),int(len(train_file)*0.1)
num_train,num_validation
```

.values to convert dataframe to numpy array

```python
#generate training data from train_file
x_train,y_train=train_file.iloc[:num_train,1:].values,train_file.iloc[:num_train,0].values
```

```python
x_test , y_test = test_file.iloc[:1000,1:].values , test_file.iloc[:1000,0].values
```

```python
x_validation,y_validation=train_file.iloc[num_train:,1:].values,train_file.iloc[num_train:,0].values
```

```python
# x--> input and y--> output
print(x_train.shape)
print(y_train.shape)
print(x_validation.shape)
print(y_validation.shape)
```

```python
#image fixing
for i in range(5):
  if (alpha.get(y_train[i]) == 'n'):
    image = x_train[i].reshape(28,28)
    image = np.fliplr(image)
    #cv2_imshow(image)
    image = np.rot90(image) #rotates the image in anti-clock wise direction
    cv2_imshow(image)
  
```

```python
# re shape

x_train = x_train.reshape(-1,28,28,1)

#cv2_imshow(x_train[4])
for i in range(79919):
 # x_train[i] = cv2.resize(x_train[i],(140,140))
  x_train[i] =np.fliplr(x_train[i])
  x_train[i] = np.rot90(x_train[i])

  
```

```python
# y reshape

y_value = np.zeros((79919,26))

for i in range(79919):
  y_value[i][y_train[i]-1] = 1 
```

```python
# x validation

x_validation = x_validation.reshape(-1,28,28,1)

for i in range(8880):
  x_validation[i] =np.fliplr(x_validation[i])
  x_validation[i] = np.rot90(x_validation[i])
  
```

```python
# y validation 

y_val = np.zeros((8880,26))

for i in range(8880):
  y_val[i][y_validation[i]-1] = 1 
```

```python
# x_test

x_testv = x_test.reshape(-1,28,28,1)

for i in range(1000):
  x_testv[i] =np.fliplr(x_validation[i])
  x_testv[i] = np.rot90(x_validation[i])
  
```

```python
#y test

y_testv = np.zeros((1000,26))

for i in range(1000):
  y_testv[i][y_test[i]-1] = 1 
```

SIMPLE CHARACTER RECOGNITION MODEL BUILDING

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers ,models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Dense,Flatten,Dropout
```

MODEL-2

```python
model=Sequential([
Conv2D(32, kernel_size=5, padding='same',activation='relu', input_shape=(28,28,1)),
MaxPool2D(pool_size= (2,2),strides=2),

Dropout(0.3),

Conv2D(64, kernel_size=5,padding='same',activation='relu'),
MaxPool2D(pool_size= (2,2),strides=2),

Dropout(0.3),

Flatten(),

Dense(128,activation='relu'),
Dropout(0.3),

model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
Dense(26, activation='softmax')])
```

```python
model.summary()
```

```python
# training the model for 10 epochs
model.fit(x_train, y_value, batch_size=128, epochs=20, verbose = 1,validation_data=(x_validation,y_val))
```

```python
model.save('SIH20epochmodel.h5')
```

```python
model.evaluate(x_train, y_value,verbose=0)
```
